```python
2. In a list of values stored in a variable called spam, how would you assign the value 'hello' as the 
third value? (Assume [2, 4, 6, 8, 10] are in spam.)
Let's pretend the spam includes the list ['a', 'b', 'c', 'd'] for the next three queries.
Ans:-
In Python, lists use zero-based indexing, meaning:

1st element → index 0
2nd element → index 1
3rd element → index 2
Given:
spam = [2, 4, 6, 8, 10]
To assign 'hello' as the third value:
spam[2] = 'hello'
Result:
spam = [2, 4, 'hello', 8, 10]
Key idea:
Index 2 corresponds to the third position in the list.

For the next questions, we’ll assume:

spam = ['a', 'b', 'c', 'd']
```

```python
3. What is the value of spam[int(int('3' * 2) / 11)]?
Ans:-
Let’s evaluate step by step in Python:

Given:
spam = ['a', 'b', 'c', 'd']
Expression:
spam[int(int('3' * 2) / 11)]
Step-by-step breakdown:
'3' * 2 → '33'
int('33') → 33
33 / 11 → 3.0
int(3.0) → 3

So the expression becomes:

spam[3]
Final Answer:
'd'
```

```python
4. What is the value of spam[-1]?
Ans:-
In Python, negative indexing accesses elements from the end of the list.

Given:
spam = ['a', 'b', 'c', 'd']
Explanation:
spam[-1] → last element
spam[-2] → second last element
So:
spam[-1]
Final Answer:
'd'
```

```python
6. What is the value of bacon.index('cat')?
Ans:-
In Python, the .index() method returns the index of the first occurrence of a value in a list.

Given:
bacon = [3.14, 'cat', 11, 'cat', True]
Expression:
bacon.index('cat')
Explanation:
'cat' appears at:
index 1 (first occurrence)
index 3 (second occurrence)
.index() always returns the first match
Final Answer:
1
```

```python
7. How does bacon.append(99) change the look of the list value in bacon?
Ans:- 
In Python, the .append() method adds a new element to the end of a list.

Given:
bacon = [3.14, 'cat', 11, 'cat', True]
Operation:
bacon.append(99)
Result:
bacon = [3.14, 'cat', 11, 'cat', True, 99]
```

```python
8. How does bacon.remove('cat') change the look of the list in bacon?
Ans:-
In Python, the .remove() method deletes the first occurrence of a specified value.

Given:
bacon = [3.14, 'cat', 11, 'cat', True]
Operation:
bacon.remove('cat')
Explanation:
'cat' appears at index 1 and 3
.remove('cat') removes only the first occurrence (index 1)
Result:
bacon = [3.14, 11, 'cat', True]
```

```python
9. What are the list concatenation and list replication operators?
Ans:- 
1. List Concatenation Operator (+)
Combines two lists into a single list
Example:
[1, 2, 3] + [4, 5]
Result:
[1, 2, 3, 4, 5]
2. List Replication Operator (*)
Repeats the list a specified number of times
Example:
['a', 'b'] * 3
Result:
['a', 'b', 'a', 'b', 'a', 'b']
```

```python
10. What is difference between the list methods append() and insert()?
Ans:-
append(x)
→ Adds x at the end of the list

lst = [1, 2]
lst.append(3)   # [1, 2, 3]

insert(i, x)
→ Adds x at a specific index i

lst = [1, 2]
lst.insert(1, 3)   # [1, 3, 2]
```

```python
11. What are the two methods for removing items from a list?
Ans:-
remove(x) → removes first occurrence of value

lst.remove(2)

pop([i]) → removes and returns item at index (default = last)

lst.pop()      # last item
lst.pop(1)     # index 1
```

```python
13. What's the difference between tuples and lists?
Ans:-
Feature	List	Tuple
Syntax	[ ]	( )
Mutability	Mutable	Immutable
Modification	Can change	Cannot change

Example:

lst = [1, 2]
tup = (1, 2)

lst[0] = 10   # allowed
tup[0] = 10   # error
```

```python
15. How do you get a list value's tuple form? How do you get a tuple value's list form?
Ans:-
In Python, you use built-in type conversion functions:

Convert list → tuple

Use tuple():

my_list = [1, 2, 3]
result = tuple(my_list)
Result:
(1, 2, 3)
Convert tuple → list

Use list():

my_tuple = (1, 2, 3)
result = list(my_tuple)
Result:
[1, 2, 3]
```

```python
16. Variables that "contain" list values are not necessarily lists themselves. Instead, what do they
contain?
Ans:-
In Python, variables don’t actually store the list itself—they store a reference (memory address) to the list object.

Explanation:

When you write:

spam = [1, 2, 3]
The list [1, 2, 3] is created somewhere in memory
spam stores a reference pointing to that list, not the list directly
Key idea:
Variables contain references to objects, not the objects themselves
Why this matters:
a = [1, 2, 3]
b = a
b.append(4)

Now:

a  # [1, 2, 3, 4]
b  # [1, 2, 3, 4]
```

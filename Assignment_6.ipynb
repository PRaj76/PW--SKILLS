```python
1. What are escape characters, and how do you use them?
Ans:-
Escape characters are special sequences used inside strings to represent characters that are otherwise difficult to include.

They start with a backslash \.

Example:
print("Hello\nWorld")

Output:

Hello
World
```

```python
2. What do the escape characters \n and \t stand for?
Ans:-
\n → newline (moves text to next line)
\t → tab (adds horizontal space)
```

```python
3. How do you include a backslash in a string?
Ans:-
Use double backslash:

print("This is a backslash: \\")
```

```python
4. Why is "Howl's Moving Castle" valid without escaping?
Ans:-
Because the string is enclosed in double quotes, the single quote ' inside does not terminate the string.

"Howl's Moving Castle"  # valid

If using single quotes:

'Howl\'s Moving Castle'  # escape needed
```

```python
5. How do you write a string of newlines without using \n?
Ans:-
Use triple-quoted strings:

print("""Line1
Line2
Line3""")
```

```python
6. Values of expressions
Ans:-
'Hello, world!'[1]     # 'e'
'Hello, world!'[0:5]   # 'Hello'
'Hello, world!'[:5]    # 'Hello'
'Hello, world!'[3:]    # 'lo, world!'
```

```python
7. Values of expressions
Ans:-
'Hello'.upper()                 # 'HELLO'
'Hello'.upper().isupper()       # True
'Hello'.upper().lower()         # 'hello'
```

```python
8. Values of expressions
Ans:-
'Remember, remember, the fifth of July.'.split()
# ['Remember,', 'remember,', 'the', 'fifth', 'of', 'July.']

'-'.join('There can only one.'.split())
# 'There-can-only-one.'
```

```python
9. Methods for string alignment
Ans:-
Right justify → .rjust(width)
Left justify → .ljust(width)
Center → .center(width)
Example:
'hi'.rjust(5)   # '   hi'
'hi'.ljust(5)   # 'hi   '
'hi'.center(5)  # ' hi  '
```

```python
10. Best way to remove whitespace
Ans:-

Use:

.strip() → both sides
.lstrip() → left side
.rstrip() → right side
Example:
'  hello  '.strip()   # 'hello'
```

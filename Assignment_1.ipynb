```python
1. In the below elements which of them are values or an expression? eg:- values can be
integer or string and expressions will be mathematical operators.
Ans:-
✅ Values (data)

These represent actual data like numbers or strings:

*'hello' → String value
*-87.8 → Floating-point value
*6 → Integer value

⚙️ Operators (used in expressions)

These are symbols used to perform operations:

*- → Subtraction operator (also unary minus)
*/ → Division operator
*+ → Addition operator

2. What is the difference between string and variable?
Ans:-
Here’s a precise distinction:

🔹 String

A string is a data type (value) used to represent text.

Written inside quotes: ' ' or " "
Example:
'hello'
"Pankaj"
It is fixed data, not a container

👉 Think: actual content

🔹 Variable

A variable is a name (identifier) used to store a value (like a string, number, etc.)

No quotes

Example:

name = "Pankaj"
name → variable
"Pankaj" → string

3. Describe three different data types.
Ans:-
🔹 1. Integer (int)

Represents whole numbers (no decimal point)

Examples:
10
-5
0

👉 Used for counting, indexing, etc.

🔹 2. Floating-point (float)

Represents decimal numbers

Examples:
3.14
-87.8
0.5

👉 Used for precise calculations (like measurements)

🔹 3. String (str)

Represents text (sequence of characters)

Written inside quotes:
'hello'
"Python"
"123" (still a string, not a number)

4. What is an expression made up of? What do all expressions do?
Ans:-
🔹 What is an expression made up of?
An expression is made up of:

Values (data) → e.g., 5, 'hello'
Variables → e.g., x, name
Operators → e.g., +, -, *, /
(Sometimes) function calls → e.g., len("abc")

👉 Example:

x + 5
x → variable
5 → value
+ → operator

🔹 What do all expressions do?

All expressions evaluate to a single value.

👉 Example:

5 + 3 → 8
'hello' + ' world' → 'hello world'
10 / 2 → 5.0

No matter how simple or complex, an expression always produces one result.

5. This assignment statements, like spam = 10. What is the difference between an
expression and a statement?
Ans:-
🔹 Expression vs Statement (clear distinction)
✅ Expression

An expression is something that evaluates to a value.

Examples:
5 + 3 → 8
'hello' + ' world' → 'hello world'
10 → 10

👉 It produces a result

✅ Statement

A statement is an instruction that Python executes.

Examples:

spam = 10
print(spam)
spam = 10 → assignment statement
It does something, but doesn’t just return a value

👉 It performs an action

🔸 Key Difference
Feature	Expression	Statement
Purpose	Produces a value	Performs an action
Output	Always returns a value	May not return anything
Example	5 + 3	spam = 10
🔹 Important Point

👉 In spam = 10:

10 → expression (value)
spam = 10 → statement
✅ Final Summary
Expression → gives a value
Statement → executes an action

6. After running the following code, what does the variable bacon contain?
bacon = 22
bacon + 1
Ans:-
Let’s analyze step by step:

bacon = 22
bacon + 1
🔹 Step 1
bacon = 22 → assigns 22 to the variable bacon
🔹 Step 2
bacon + 1 → this is an expression
It evaluates to 23

👉 But important point:
This result is not stored anywhere

✅ Final Answer:

The variable bacon still contains 22

🔹 Why?

Because:

bacon + 1 does not change the variable
It only calculates the value temporarily
🔸 To change the value, you must write:
bacon = bacon + 1

Now bacon will become 23

7. What should the values of the following two terms be?
'spam' + 'spamspam'
'spam' * 3
Ans:-
Let’s evaluate both expressions:

🔹 1. 'spam' + 'spamspam'
+ → concatenation (joining strings)

Result:

'spamspamspam'
🔹 2. 'spam' * 3
* → repetition operator for strings

Result:

'spamspamspam'

✅ Final Answer:
'spam' + 'spamspam' → 'spamspamspam'
'spam' * 3 → 'spamspamspam'


8. Why is eggs a valid variable name while 100 is invalid?
Ans:-
🔹 Why eggs is valid and 100 is invalid
✅ eggs (Valid variable name)
Starts with a letter ✅
Contains only letters ✅
Follows Python naming rules ✅

👉 So, it can be used as a variable:

eggs = 10

9. What three functions can be used to get the integer, floating-point number, or string
version of a value?
Ans:-
🔹 Three functions for type conversion in Python

To convert values into different data types, Python provides these built-in functions:

✅ 1. int()
Converts a value to an integer

Examples:

int(3.7)      # → 3
int("10")     # → 10
✅ 2. float()
Converts a value to a floating-point number

Examples:

float(5)      # → 5.0
float("3.14") # → 3.14
✅ 3. str()
Converts a value to a string

Examples:

str(100)      # → "100"
str(3.5)      # → "3.5"

10. Why does this expression cause an error? How can you fix it?
'I have eaten ' + 99 + ' burritos.'
Ans:-
🔹 Why this causes an error
'I have eaten ' + 99 + ' burritos.'

👉 Error occurs because:

'I have eaten ' → string
99 → integer

Python cannot concatenate (join) a string and an integer using +.

🔹 How to fix it

You need to convert the integer to a string using str().

'I have eaten ' + str(99) + ' burritos.'

✅ Output:

'I have eaten 99 burritos.'
🔸 Alternative (better way 👍)

Using formatted strings (f-strings):

f'I have eaten {99} burritos.'
✅ Final Answer
Error happens due to string + integer mismatch
Fix by converting number using str() or using f-strings
```

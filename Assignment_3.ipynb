```python
1. Why are functions advantageous in programs?
Ans:-Functions provide:
>Code reusability (write once, use many times)
>Modularity (break program into smaller parts)
>Better readability & maintenance
>Avoid repetition (DRY principle)

2. When does the code in a function run?
And:-The code runs only when the function is called, not when it is defined.                  

3. What statement creates a function?

Ans:-The def statement is used.
Example:
def my_function():
    pass

4. Difference between a function and a function call?
Ans:-
>Function → Definition (blueprint of code)
>Function call → Execution of that function

Example:

def greet():   # function
    print("Hello")

greet()        # function call

5. Global and Local scopes in Python?
Ans:-
>Global scope → Only one per program
>Local scope → Created every time a function is called

6. What happens to local variables after function returns?
Ans:-They are destroyed (deleted from memory) after the function finishes execution.

7. Concept of return value? Can it be used in expression?
Ans:-
A return value is the value a function sends back using return.
Yes, it can be used in expressions.

Example:

def add(a, b):
    return a + b

result = add(2, 3) * 2   # valid

8. Return value if no return statement?
Ans:-The function returns None by default.


9. How to make a function variable refer to global variable?
Ans:-
Use the global keyword.

Example:

x = 10

def func():
    global x
    x = 20

10. Data type of None?
Ans:-The data type is NoneType.

11. What does import areallyourpetsnamederic do?
Ans:-
It imports a module named areallyourpetsnamederic (if it exists).
Otherwise, it will raise an ImportError.

12. Calling bacon() from spam module?
Ans:-
import spam
spam.bacon()

13. How to prevent program from crashing on error?
Ans:-
Use exception handling with try and except.

```

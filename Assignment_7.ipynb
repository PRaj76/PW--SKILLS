```python
1. Feature responsible for generating Regex objects
Ans:-
The re.compile() function creates Regex (pattern) objects.
```

```python
2. Why raw strings are used in Regex
Ans:-
Raw strings (r'') prevent Python from interpreting backslashes as escape characters, so patterns like \d, \n are passed correctly to the regex engine.
```

```python
3. Return value of search()
Ans:-
The search() method returns a Match object if a match is found, otherwise None.
```

```python
4. Getting matched strings from Match object
Ans:-
Use:

group() → entire match
group(n) → specific group
groups() → tuple of all groups
```

```python
5. Groups in r'(\d\d\d)-(\d\d\d-\d\d\d\d)'
Ans:-
Group 0 → Entire match (e.g., 415-555-4242)
Group 1 → First part (415)
Group 2 → Second part (555-4242)
```

```python
6. Matching literal parentheses and periods
Ans:-
Escape them using backslash:

\(, \)
\.
```

```python
7. findall() return types
Ans:-
Returns list of strings → if no groups
Returns list of tuples → if groups () are present
```

```python
8. Meaning of | in expressions
Ans:-
Acts as OR operator (matches either pattern)
```

```python
9. Meaning of ? in regex
Ans:-
Matches 0 or 1 occurrence of the preceding group (optional match)
```

```python
10. Difference between + and *
Ans:-
+ → 1 or more
* → 0 or more
```

```python
11. Difference between {4} and {4,5}
Ans:-
{4} → exactly 4 times
{4,5} → between 4 and 5 times
```

```python
12. Meaning of \d, \w, \s
Ans:-
\d → digits (0–9)
\w → word characters (letters, digits, underscore)
\s → whitespace (space, tab, newline)
```

```python
13. Meaning of \D, \W, \S
Ans:-
\D → non-digits
\W → non-word characters
\S → non-whitespace
```

```python
14. Difference between .* and .*?
Ans:-
.* → greedy (matches maximum)
.*? → non-greedy/lazy (matches minimum)
```

```python
15. Matching numbers and lowercase letters
Ans:-
[a-z0-9]
```

```python
16. Making regex case-insensitive
Ans:-

Use:

re.compile(pattern, re.IGNORECASE)
```

```python
17. Meaning of .
Ans:-
Normally → matches any character except newline
With re.DOTALL → matches any character including newline
```

```python
18. Output of substitution
Ans:-
numReg = re.compile(r'\d+')
numReg.sub('X', '11 drummers, 10 pipers, five rings, 4 hen')

Output:

X drummers, X pipers, five rings, X hen
```

```python
19. Use of re.VERBOSE
Ans:-

Allows:

Multi-line regex
Comments inside regex
Better readability
```

```python
20. Regex for comma-separated numbers
Ans:-
^\d{1,3}(,\d{3})*$
```

```python
21. Regex for full name ending with Watanabe
Ans:-
^[A-Z][a-zA-Z]*\sWatanabe$
```

```python
22. Regex for specific sentence pattern
Ans:-
^(Alice|Bob|Carol)\s(eats|pets|throws)\s(apples|cats|baseballs)\.$

With case-insensitive flag:

re.compile(pattern, re.IGNORECASE)
```

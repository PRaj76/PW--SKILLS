```python
1. What does an empty dictionary's code look like?
An:-
An empty dictionary in Python is written using curly braces:

{}

Alternatively, you can also create it using the built-in constructor:

dict()
Key Point

Both create an empty dictionary object, but {} is the more commonly used and concise syntax.
```

```python
2. What is the value of a dictionary value with the key 'foo' and the value 42?
Ans:-
A dictionary with the key 'foo' and the value 42 is written as:

{'foo': 42}
Explanation
'foo' → key
42 → value
: → separates key and value
{} → defines the dictionary

If you access the value using the key:

my_dict = {'foo': 42}
print(my_dict['foo'])

Output:

42
```

```python
3. What is the most significant distinction between a dictionary and a list?
Ans:-
Dictionary
Stores data as key–value pairs
Accessed using keys
my_dict = {'a': 1, 'b': 2}
print(my_dict['a'])  # 1
List
Stores data as an ordered sequence of items
Accessed using index positions (0, 1, 2, …)
my_list = [1, 2, 3]
print(my_list[0])  # 1
```

```python
4. What happens if you try to access spam['foo'] if spam is {'bar': 100}?
Ans:-
If you try to access:

spam = {'bar': 100}
spam['foo']

Python will raise a KeyError because the key 'foo' does not exist in the dictionary.

Result:
KeyError: 'foo'
Why?
The dictionary only has the key 'bar'
'foo' is not present, so Python cannot return a value
Safer Alternative

Use .get() to avoid an error:

spam.get('foo')

This will return:

None

Or with a default value:

spam.get('foo', 0)  # returns 0 instead of error
```

```python
5. If a dictionary is stored in spam, what is the difference between the expressions 'cat'in spam and 'cat' in spam.keys()?
Ans:-
There is no practical difference between:

'cat' in spam

and

'cat' in spam.keys()
```

```python
6. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.values()?
Ans:-
The difference is what is being checked:

1. 'cat' in spam
Checks whether 'cat' exists as a key in the dictionary
2. 'cat' in spam.values()
Checks whether 'cat' exists as a value in the dictionary
Example
spam = {'cat': 42, 'dog': 100}

print('cat' in spam)          # True  (it's a key)
print('cat' in spam.values()) # False (not a value)

print(42 in spam.values())    # True  (42 is a value)
```

```python
7. What is a shortcut for the following code?
if 'color' not in spam:
spam['color'] = 'black'
Ans:-
The shortcut for that code is to use the setdefault() method:

spam.setdefault('color', 'black')
Explanation
If 'color' already exists, nothing changes
If 'color' does not exist, it is added with value 'black'
Equivalent to:
if 'color' not in spam:
    spam['color'] = 'black'
```

```python
8. How do you "pretty print" dictionary values using which module and function?
Ans:-
To “pretty print” dictionary values in Python, use the pprint module and its pprint() function.

Example
import pprint

spam = {'name': 'Pankaj', 'age': 25, 'skills': ['Python', 'SQL', 'ML']}

pprint.pprint(spam)
Explanation
pprint → stands for pretty print
pprint.pprint() → prints data structures (like dictionaries) in a well-formatted, readable way
Shortcut (optional)
from pprint import pprint

pprint(spam)
```
